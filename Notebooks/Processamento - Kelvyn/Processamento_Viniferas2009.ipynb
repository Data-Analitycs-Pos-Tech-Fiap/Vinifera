### Plotar gráfico por processamento em 2023

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mtick

# Filtrando os dados para valores acima de 5 milhões
dados_filtrados = dados[dados['2023'] > 500000]

# Ordenando os dados do maior para o menor
dados_filtrados = dados_filtrados.sort_values(by='2023', ascending=False)

# Ajuste do tamanho da figura
plt.figure(figsize=(12, 6))

# Criação do gráfico de barras horizontal
grafico = sns.barplot(x='2023', y='control', data=dados_filtrados, palette='viridis')

# Formatação do eixo X
plt.gca().xaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'R${x:,.2f}'.replace(',', 'x').replace('.', ',').replace('x', '.')))

# Ajuste dos rótulos do eixo Y
plt.yticks(rotation=0, fontsize=8)

# Título do gráfico
plt.title('Custo de processamento por tipo de uva', fontsize=16)

# Definindo a fonte global do gráfico
plt.rcParams.update({'font.family': 'serif'})

# Exibição do gráfico
plt.show()
```

## Plotar dados em barras entre 2017 e 2023

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mtick

# Filtrando os dados para valores acima de 5 milhões em qualquer ano
dados_filtrados = dados[(dados['2016'] > 500000) | (dados['2017'] > 500000) | (dados['2018'] > 500000) | (dados['2020'] > 500000) | (dados['2021'] > 500000) | (dados['2023'] > 500000)]

# Ordenando os dados com base no ano de 2023
dados_filtrados = dados_filtrados.sort_values(by='2023', ascending=False)

# Ajuste do tamanho da figura
plt.figure(figsize=(12, 6))

# Criação de gráfico de barras agrupadas para múltiplos anos
grafico = sns.barplot(x='2023', y='control', data=dados_filtrados, color='blue', label='2023')
grafico = sns.barplot(x='2022', y='control', data=dados_filtrados, color='green', label='2022')
grafico = sns.barplot(x='2021', y='control', data=dados_filtrados, color='orange', label='2021')
grafico = sns.barplot(x='2021', y='control', data=dados_filtrados, color='red', label='2020')
grafico = sns.barplot(x='2021', y='control', data=dados_filtrados, color='yellow', label='2018')
grafico = sns.barplot(x='2021', y='control', data=dados_filtrados, color='purple', label='2017')
# Formatação do eixo X
plt.gca().xaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'R${x:,.2f}'.replace(',', 'x').replace('.', ',').replace('x', '.')))

# Ajuste dos rótulos do eixo Y
plt.yticks(rotation=0, fontsize=8)

# Título do gráfico
plt.title('Custo de processamento por tipo de uva (2017-2023)', fontsize=16)

# Exibição da legenda
plt.legend(title='Ano')

# Definindo a fonte global do gráfico
plt.rcParams.update({'font.family': 'serif'})

# Exibição do gráfico
plt.show()
```

## Plotar Gráficos de linhas entre 2016 a 2023 Derretendo os dados pra que estejam em 3 colunas

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mtick

# Transformando os dados para o formato "long"
dados_melted = dados_filtrados.melt(id_vars='control', value_vars=['2016', '2017', '2018', '2020', '2021', '2023'], var_name='Ano', value_name='Custo')

# Ajuste do tamanho da figura
plt.figure(figsize=(12, 6))

# Plotando gráfico de linhas para cada ano
sns.lineplot(x='Ano', y='Custo', hue='control', data=dados_melted, marker='o')

# Formatação do eixo Y
plt.gca().yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'R${x:,.2f}'.replace(',', 'x').replace('.', ',').replace('x', '.')))

# Título do gráfico
plt.title('Evolução do Custo de Processamento por Tipo de Uva (2017-2023)', fontsize=16)

# Ajuste dos rótulos do eixo X e Y
plt.xticks(rotation=45)
plt.yticks(rotation=0, fontsize=8)

# Exibição da legenda
plt.legend(title='Tipo de Uva')

# Exibição do gráfico
plt.show()
```

```python
import pandas as pd

# Supondo que 'dados' seja o DataFrame original

# Definindo o intervalo dos últimos 15 anos
anos_recentes = list(range(2009, 2024))  # Ajuste conforme necessário
anos_recentes = [str(ano) for ano in anos_recentes]  # Convertendo os anos para string

# Separar as colunas adicionais
colunas_adicionais = ['control', 'id']  # Adicione outras colunas adicionais se houver
colunas_anos = [col for col in dados.columns if col in anos_recentes]

# Filtrar apenas as colunas de anos
dados_anos = dados[colunas_anos]

# Convertendo as colunas para numérico
dados_anos = dados_anos.apply(pd.to_numeric, errors='coerce')

# Adicionar as colunas adicionais de volta
dados_completos = dados[colunas_adicionais + colunas_anos]

# Filtrar linhas onde o custo de processamento é maior que 5 milhões em pelo menos um dos anos
dados_filtrados = dados_completos[dados_anos.max(axis=1) > 5000000]

# Exibir o DataFrame resultante
print(dados_filtrados)
```

```python
dados_filtrados.head()
```

```python
dados_filtrados.rename(columns={'control': 'Tipo_de_Uva'}, inplace=True)
```

```python
dados_filtrados.set_index('Tipo_de_Uva', inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mtick
# Reset the index to make 'Tipo_de_Uva' a column again
dados_filtrados.reset_index(inplace=True)

# Now you can melt the DataFrame
dados_melted = dados_filtrados.melt(id_vars='Tipo_de_Uva', value_vars=['2016', '2017', '2018', '2020', '2021', '2023'], var_name='Ano', value_name='Custo')

# Ajuste do tamanho da figura
plt.figure(figsize=(12, 6))

# Plotando gráfico de linhas para cada ano
sns.lineplot(x='Ano', y='Custo', hue='Tipo_de_Uva', data=dados_melted, marker='o')

# Formatação do eixo Y
plt.gca().yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'R${x:,.2f}'.replace(',', 'x').replace('.', ',').replace('x', '.')))

# Título do gráfico
plt.title('Evolução do Custo de Processamento por Tipo de Uva (2017-2023)', fontsize=16)

# Ajuste dos rótulos do eixo X e Y
plt.xticks(rotation=45)
plt.yticks(rotation=0, fontsize=8)

# Exibição da legenda
plt.legend(title='Tipo de Uva')

# Exibição do gráfico
plt.show()
```
