# AnÃ¡lise dos Custos de Processamento de Vinhos ao Longo dos Anos
Nesta anÃ¡lise, exploramos os custos de processamento de vinhos ao longo dos anos, utilizando dados histÃ³ricos disponÃ­veis. O objetivo Ã© identificar tendÃªncias e padrÃµes nos custos ao longo do tempo, o que pode fornecer insights valiosos para otimizaÃ§Ã£o de processos e tomada de decisÃµes estratÃ©gicas.

SeleÃ§Ã£o dos Anos Relevantes
Primeiramente, selecionamos os anos relevantes para a anÃ¡lise. Isso nos permite focar nos perÃ­odos de interesse e garantir que estamos analisando dados consistentes e significativos. A seleÃ§Ã£o dos anos Ã© crucial para entender como os custos evoluÃ­ram ao longo do tempo.

CÃ¡lculo do Total por Ano
Em seguida, calculamos o total de custos de processamento para cada ano. Este cÃ¡lculo nos fornece uma visÃ£o agregada dos custos anuais, facilitando a comparaÃ§Ã£o entre diferentes perÃ­odos. Para tornar os valores mais manejÃ¡veis e compreensÃ­veis, os custos sÃ£o convertidos para milhÃµes de reais.

VisualizaÃ§Ã£o dos Dados
Para visualizar a tendÃªncia dos custos ao longo dos anos, criamos um grÃ¡fico de linha. Este tipo de grÃ¡fico Ã© ideal para mostrar mudanÃ§as ao longo do tempo, permitindo identificar facilmente aumentos, diminuiÃ§Ãµes e perÃ­odos de estabilidade nos custos de processamento.

InterpretaÃ§Ã£o do GrÃ¡fico
O grÃ¡fico resultante mostra o total de custos de processamento de vinhos ao longo dos anos, com os anos no eixo x e os custos totais no eixo y. Os pontos de dados sÃ£o destacados com marcadores para facilitar a visualizaÃ§Ã£o. A grade adicionada ao grÃ¡fico ajuda na leitura dos valores, e os rÃ³tulos dos eixos fornecem contexto adicional, indicando que os custos estÃ£o em milhÃµes de reais.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Selecionando os anos relevantes para calcular o total por ano
anos = viniferas.columns[39:].astype(str)
total_por_ano = viniferas[anos].sum(numeric_only=True) / 1_000_000  # Convertendo para milhÃµes

# GrÃ¡fico de linha para o total de custos ao longo dos anos
plt.figure(figsize=(10, 5))
plt.plot(total_por_ano.index, total_por_ano.values, marker='o')
plt.title('Total de Custos de Processamento de Vinhos ao Longo dos Anos (em MilhÃµes)')
plt.xlabel('Ano')
plt.ylabel('Custo Total (R$ MilhÃµes)')  # Ajuste no rÃ³tulo para refletir a unidade
plt.grid()
plt.xticks(rotation=45)  # Opcional: rotacionar os rÃ³tulos do eixo x
plt.show()

```

Nesta anÃ¡lise, exploramos os custos de processamento de diferentes tipos de uva ao longo dos anos, com foco nos cinco tipos que apresentam os maiores custos. O objetivo Ã© identificar tendÃªncias e padrÃµes nesses custos, oferecendo insights para a otimizaÃ§Ã£o de processos e a tomada de decisÃµes estratÃ©gicas.

Inicialmente, calculamos o custo mÃ©dio de processamento para cada tipo de uva, o que nos fornece uma visÃ£o consolidada e facilita a comparaÃ§Ã£o entre elas. Em seguida, selecionamos os cinco tipos de uva com os maiores custos mÃ©dios de processamento, permitindo uma anÃ¡lise focada nas variedades mais relevantes em termos de impacto financeiro.

O grÃ¡fico gerado exibe a evoluÃ§Ã£o dos custos de processamento ao longo dos anos, com o eixo x representando os anos e o eixo y os custos, expressos em milhÃµes de reais. Linhas de referÃªncia horizontais e verticais foram adicionadas para facilitar a leitura dos valores e a identificaÃ§Ã£o de padrÃµes. RÃ³tulos e legendas tambÃ©m fornecem contexto adicional, diferenciando os tipos de uva analisados.

AtravÃ©s desta anÃ¡lise, Ã© possÃ­vel observar como os custos de processamento variam entre os diferentes tipos de uva ao longo dos anos. Essa visualizaÃ§Ã£o se mostra uma ferramenta eficaz para identificar tendÃªncias e embasar decisÃµes informadas com base nos dados histÃ³ricos. A anÃ¡lise contÃ­nua desses dados pode contribuir para a otimizaÃ§Ã£o dos processos produtivos e a reduÃ§Ã£o de custos no futuro.

```python
# Transpondo o DataFrame para facilitar a visualizaÃ§Ã£o
viniferas_melted = viniferas.melt(id_vars=['Tipo_Uva'], var_name='Ano', value_name='Custo')

# Convertendo a coluna 'Custo' para tipo numÃ©rico
viniferas_melted['Custo'] = pd.to_numeric(viniferas_melted['Custo'], errors='coerce')

# Calculando o custo mÃ©dio por tipo de uva
custo_medio = viniferas_melted.groupby('Tipo_Uva')['Custo'].mean().reset_index()

# Obtendo os 5 tipos de uva com os maiores custos
top5_custos = custo_medio.nlargest(5, 'Custo')

# Filtrando os dados para incluir apenas os tipos de uva que estÃ£o entre os 5 maiores custos
viniferas_top5 = viniferas_melted[viniferas_melted['Tipo_Uva'].isin(top5_custos['Tipo_Uva'])]

# Convertendo o custo para milhÃµes
viniferas_top5['Custo'] = viniferas_top5['Custo'] / 1_000_000

# AnÃ¡lise de TendÃªncias de Custos para os 5 tipos de uva com maiores custos
plt.figure(figsize=(24, 12))

# Criando o grÃ¡fico de linha
ax = sns.lineplot(data=viniferas_top5, x='Ano', y='Custo', hue='Tipo_Uva', marker='o')

# Adicionando linhas de referÃªncia no fundo (pontilhadas e claras)
for y in range(0, int(viniferas_top5['Custo'].max()), 5):  # Ajuste o intervalo conforme necessÃ¡rio
    ax.axhline(y=y, color='lightgray', linestyle='--', linewidth=0.8)

# Adicionando linhas verticais pontilhadas para cada ano
for ano in viniferas_top5['Ano'].unique():
    ax.axvline(x=ano, color='lightgray', linestyle='--', linewidth=0.8)

# TÃ­tulo e rÃ³tulos
plt.title('Custos de Processamento por Tipo de Uva (Top 5 com Maiores Custos)')
plt.xticks(rotation=45)
plt.ylabel('Custo de Processamento (em MilhÃµes)')
plt.legend(title='Tipo de Uva')

plt.show()

```
