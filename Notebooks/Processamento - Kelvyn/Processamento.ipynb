Seleção dos Anos Relevantes

Cálculo do Total por Ano

Para visualizar a tendência dos custos ao longo dos anos, criamos um gráfico de linha. Este tipo de gráfico é ideal para mostrar mudanças ao longo do tempo, permitindo identificar facilmente aumentos, diminuições e períodos de estabilidade nos custos de processamento.

Interpretação do Gráfico
O gráfico resultante mostra o total de custos de processamento de vinhos ao longo dos anos, com os anos no eixo x e os custos totais no eixo y. Os pontos de dados são destacados com marcadores para facilitar a visualização. A grade adicionada ao gráfico ajuda na leitura dos valores, e os rótulos dos eixos fornecem contexto adicional, indicando que os custos estão em milhões de reais.

# Análise dos Custos de Processamento por Tipo de Uva

Nesta análise, exploramos os custos de processamento de diferentes tipos de uva ao longo dos anos, com foco nos cinco tipos que apresentam os maiores custos. O objetivo é identificar tendências e padrões nesses custos, oferecendo insights para a otimização de processos e a tomada de decisões estratégicas.

Inicialmente, calculamos o custo médio de processamento para cada tipo de uva, o que nos fornece uma visão consolidada e facilita a comparação entre elas. Em seguida, selecionamos os cinco tipos de uva com os maiores custos médios de processamento, permitindo uma análise focada nas variedades mais relevantes em termos de impacto financeiro.

O gráfico gerado exibe a evolução dos custos de processamento ao longo dos anos, com o eixo x representando os anos e o eixo y os custos, expressos em milhões de reais. Linhas de referência horizontais e verticais foram adicionadas para facilitar a leitura dos valores e a identificação de padrões. Rótulos e legendas também fornecem contexto adicional, diferenciando os tipos de uva analisados.

Através desta análise, é possível observar como os custos de processamento variam entre os diferentes tipos de uva ao longo dos anos. Essa visualização se mostra uma ferramenta eficaz para identificar tendências e embasar decisões informadas com base nos dados históricos. A análise contínua desses dados pode contribuir para a otimização dos processos produtivos e a redução de custos no futuro.

```python
# Transpondo o DataFrame para facilitar a visualização
viniferas_melted = viniferas.melt(id_vars=['Tipo_Uva'], var_name='Ano', value_name='Custo')

# Convertendo a coluna 'Custo' para tipo numérico
viniferas_melted['Custo'] = pd.to_numeric(viniferas_melted['Custo'], errors='coerce')

# Calculando o custo médio por tipo de uva
custo_medio = viniferas_melted.groupby('Tipo_Uva')['Custo'].mean().reset_index()

# Obtendo os 5 tipos de uva com os maiores custos
top5_custos = custo_medio.nlargest(5, 'Custo')

# Filtrando os dados para incluir apenas os tipos de uva que estão entre os 5 maiores custos
viniferas_top5 = viniferas_melted[viniferas_melted['Tipo_Uva'].isin(top5_custos['Tipo_Uva'])]

# Convertendo o custo para milhões
viniferas_top5['Custo'] = viniferas_top5['Custo'] / 1_000_000

# Análise de Tendências de Custos para os 5 tipos de uva com maiores custos
plt.figure(figsize=(24, 12))

# Criando o gráfico de linha
ax = sns.lineplot(data=viniferas_top5, x='Ano', y='Custo', hue='Tipo_Uva', marker='o')

# Adicionando linhas de referência no fundo (pontilhadas e claras)
for y in range(0, int(viniferas_top5['Custo'].max()), 5):  # Ajuste o intervalo conforme necessário
    ax.axhline(y=y, color='lightgray', linestyle='--', linewidth=0.8)

# Adicionando linhas verticais pontilhadas para cada ano
for ano in viniferas_top5['Ano'].unique():
    ax.axvline(x=ano, color='lightgray', linestyle='--', linewidth=0.8)

# Título e rótulos
plt.title('Custos de Processamento por Tipo de Uva (Top 5 com Maiores Custos)')
plt.xticks(rotation=45)
plt.ylabel('Custo de Processamento (em Milhões)')
plt.legend(title='Tipo de Uva')

plt.show()

```
